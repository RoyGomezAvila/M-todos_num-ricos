# **Cadenas de Markov**
La cadena de Markov lleva el nombre del matemático ruso Andrey Markov (1856 - 1922). Una Cadena de Markov se define como un proceso estocástico que posee la Propiedad de Markov.


-----------------------------------------------------------------------
*Tenemos un proceso que va de un estado a otro, el nuevo estado depende solo de la posición actual y no de las posiciones históricas. Esto también se conoce como sin memoria.* 

------------------------------------------------------------------------


Algunos de los ejemplos del uso son modelar **el precio de un activo**, **la temperatura**, **el cambio de la altura del paisaje sobre el nivel del mar**, **el crecimiento de la población**, el tipo de cambio de las tasas de cambio (como las criptomonedas, por ejemplo), **PageRank** se basa en el proceso de Markov. https://en.wikipedia.org/wiki/PageRank

```python
#Aplication 
import numpy as np
import matplotlib.pyplot as plt

N=10
p=1/4
q=2/3
B=np.zeros(N)
C=np.zeros(N)
B[0]=3/5
C[0]=1-B[0]

for n in range(N-1):

  B[n+1]=p*B[n]+(1-q)*C[n]
  C[n+1]=q*C[n]+(1-p)*B[n]

plt.figure(figsize=(12,6)) #Tamaño de gráfica
plt.plot(B,label="B")
plt.plot(C,label="C")
plt.legend(loc=1)
plt.show()

print("el valor final de A es",B[-1])
print("el valor final de B es",C[-1])
```

# **EJEMPLO CONCRETO DE CADENA DE MARKOV**

Considere el siguiente sistema 

\begin{eqnarray}
pR_n &+(1-q)S_n&&=R_{n+1}\\
(1-p)R_n &+qS_n&&=S_{n+1}\\
R_{n} &+ S_{n} &&= N 
\end{eqnarray}


Tomando limites tenemos:

\begin{eqnarray}
pR &+(1-q)S&&=R\\
R &+ S &&= N 
\end{eqnarray}

Debido a que la ecuación $(1-p)R+qS=S$ proporciona el mismo conjunto solución de la primera ecuación.

Reescribiendo la primera ecuación como $(p-1)R+(1-q)S=0$ y multiplicar por $-(p-1)$ la segunda ecuación y sumarla con esta nueva versión de la primera, obtenemos:
$(2-p-q)S=-(p-1)N$, es decir $S=\frac{(1-p)N}{(2-p-q)}$.

Sustituyendo este valor en la segunda ecuación, se sigue:
\begin{eqnarray}
\begin{split}
R &= N -S\\
&= N-\frac{(1-p)N}{(2-p-q)}\\
&= N\frac{1-q}{(2-p-q)}
\end{split}
\end{eqnarray}

Al reemplazar, para $p=0.8$, $q=0.5$ y $N=1000$, se obtiene

Para $R$:

\begin{eqnarray}
\begin{split}
R &= N\frac{1-q}{(2-p-q)}\\
&= 1000\frac{1-0.5}{(2-0.8-0.5)}\\
&= 1000\frac{0.5}{0.7}\\
&= 714.285714286\\
\end{split}
\end{eqnarray}

y para $S$:

\begin{eqnarray}
\begin{split}
S &= N\frac{1-p}{(2-p-q)}\\
&= 1000\frac{1-0.8}{(2-0.8-0.5)}\\
&= 1000\frac{0.2}{0.7}\\
&= 285.714285714\\
\end{split}
\end{eqnarray}



```python
#Aplication 
import numpy as np
import matplotlib.pyplot as plt

N=10
p=0.8
q=0.5
B=np.zeros(N)
C=np.zeros(N)
B[0]=100
C[0]=1000-B[0]

for n in range(N-1):

  B[n+1]=p*B[n]+(1-q)*C[n]
  C[n+1]=q*C[n]+(1-p)*B[n]

plt.figure(figsize=(12,6)) #Tamaño de gráfica
plt.plot(B,'b',label="Equipo Azúl")
plt.plot(C,'r',label="Equipo Rojo")
plt.legend(loc=1)
plt.show()

print("el valor final de A es",B[-1])
print("el valor final de B es",C[-1])
```
