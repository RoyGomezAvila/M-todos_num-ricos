# **FRACCIONES PARCIALES Y OPERAR**

Tratamos a continuación simplificación, esto es la reescritura
de fracciones mas reducidas y operaciones algebraicas. Las funciones `apart` y `together`,lo tratan respectivamente, con lo que 
reescriben una fracción como fracción parcial y combinar fracciones parciales en una sola fracción, observemos lo siguiente:

```python
from sympy import *
x, y= symbols('x , y')
```

```python
apart(1/(x**2+3*x+2), x)
```

```python
together(1/(y*x+y)+1/(1+x))
```

```python
cancel(y/(y*x+y))
```

# **ACERCA DE ALGEBRA LINEAL**


En **SymPy** podemos representar vectores y matrices simbólicamente empleando la Clase de matrices incorporada en esta biblioteca, cuyos elementos a su vez pueden ser representados por números, símbolos o incluso expresiones simbólicas arbitrarias. Veamos:

```python
a, b, c, d = symbols('a, b, c, d')
```

```python
M = Matrix([[a, b], [c, d]])
```

```python
M
```

```python
M*M.transpose()
```

```python
x=Matrix(symbols("x_1, x_2"))
```

```python
M*x
```

```python
b =Matrix(symbols("b_1, b_2"))
```

```python
b
```

```python
M.det()
```

```python
x = M.solve(b)
x
```

```python
x = M.LUsolve(b)
x
```

```python
x = M.inv()*b
x
```

### **Lanzamiento de dos dados**

```python
X, Y = Die('X', 6), Die('Y', 6)  
density(X).dict
```
