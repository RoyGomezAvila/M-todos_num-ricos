### **Ejercicio Sympy y Scipy**

1. Emplea la librería de Sympy para resolver la siguiente ecuación diferencial $$y''+y=\mbox{sen}(x).$$

2. Emplea la librería de Scipy para resolver la ecuacion diferencial dicha.

3.  Emplea las condiciones iniciales $y(0)=1$ e $y'(0)=0$ para determinar la solución particular del problema y grafica su solución.

```python
#Solución 1
from sympy import *
x, y, z, t = symbols('x y z t')
k, m, n = symbols('k m n', integer=True)
f, g, h = symbols('f g h', cls=Function)
```

```python
dsolve(f(x).diff(x, 2)+f(x)-sin(x), f(x))
```

```python
#Solución 2
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt

def w(y,x):
  return np.array([y[1], -y[0]+np.sin(x)])
```

```python
x=np.linspace(0,100)
w0=np.array([1.0,0.0])#Condiciones iniciales.
```

```python
sol= integrate.odeint(w, w0, x)
print(sol)
```

```python
plt.figure(figsize=(10,5))

plt.plot(x,sol[:,0])
plt.xlabel("VARIABLE INDEPENDIENTE",fontsize=13)
plt.ylabel("'VARIABLE DEPENDIENTE",fontsize=13)
plt.title("SOLUCIÓN NUMÉRICA DADA POR SCIPY")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
from scipy import integrate

def u(y,x):
  return np.array([y[1], y[0]+np.exp(x)])
```

```python
x=np.linspace(0,10)
u0=np.array([0.0,1.0])
```

```python
solu= integrate.odeint(u, u0, x)
print(solu)
```

```python
plt.figure(figsize=(9,6))

plt.plot(x,solu[:,0])
plt.xlabel("VARIABLE INDEPENDIENTE",fontsize=13)
plt.ylabel("VARIABLE DEPENDIENTE",fontsize=13)
plt.title("SOLUCIÓN NUMÉRICA")
plt.legend()
plt.grid()
plt.show()
```
