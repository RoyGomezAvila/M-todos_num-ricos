*    $x_0= 0$, para cualqiuer valor de $r$.



```
import numpy as np
import matplotlib.pyplot as plt
import random

n=np.linspace(0,50,50)
r=random.random()+random.randint(-10**6,10**6)
def logistica(x):
    x=np.empty(50)
    x[0]=3
    for i in range(49):
      x[i+1]=r*(1-x[i])*x[i]         
    return x

p1=logistica(n)

plt.plot(n,p1,'g-',label="Información deterministica")

plt.xlabel("Tiempo discreto n",fontsize=15)
plt.ylabel("Término n-ésimo",fontsize=15)
plt.title("MODELO LOGÍSTICO")

plt.legend()
plt.show()
print("Este es el gráfico con x_0=",0,"y valor r=",r)
```

```
import matplotlib.pyplot as plt
import numpy as np


def Mapalogistico():
    mu = np.arange(0, 4, 0.0001)
    x = 0.2  # Valor inicial
    iteracion = 1000  # Número de iteraciones sin salida
    ultimo = 100  # Número de iteración para dibujar la salida
    for i in range(iteracion+ultimo):
        x = mu * x * (1 - x)
        if i >= iteracion:
            plt.plot(mu, x, ',c', alpha=0.25)
    plt.title("DIAGRAMA DE BIFURCACIÓN")         
    plt.show()
Mapalogistico()
```
